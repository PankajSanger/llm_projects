#Importing scraper

```python
from scraper import website_scrape
import time

message = "summaries my website content. Don't consider the urls or headers. Tell me to the point"
content = (website_scrape("https://edwarddonner.com/2024/11/13/llm-engineering-resources/"))
```

#Sending API request to llama2 model locally to summaries website

```python
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1',
    api_key='ollama',
)

response = client.chat.completions.create(
    model="llama2",
    messages=[
        {"role":"system","content":"Talk like a professional"},
        {"role":"user","content":message + content}
    ]
)

print(response.choices[0].message.content)
```
